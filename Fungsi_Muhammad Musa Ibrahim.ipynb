Problem (2) Validasi Sandi

```python
def validasi_sandi(sandi):
    # Cek panjang sandi
    if len(sandi) < 8:
        return False, "Sandi harus memiliki minimal 8 karakter."

    # Cek apakah ada minimal satu huruf besar
    if not any(char.isupper() for char in sandi):
        return False, "Sandi harus mengandung minimal satu huruf besar."

    # Cek apakah ada minimal satu angka
    if not any(char.isdigit() for char in sandi):
        return False, "Sandi harus mengandung minimal satu angka."

    # Jika semua syarat terpenuhi
    return True, "Sandi valid."

  # Contoh Penggunaan
sandi = "Fitri123"
hasil, pesan = validasi_sandi(sandi)
print(hasil, pesan)

sandi2 = "fitrii123"
hasil2, pesan2 = validasi_sandi(sandi2)
print(hasil2, pesan2)

sandi3 = "FITRI"
hasil3, pesan3 = validasi_sandi(sandi3)
print(hasil3, pesan3)
```

Problem(2) Determinan Matriks

```python
def determinant_2x2(matrix):
    # Validasi input matriks 2x2
    if len(matrix) != 2 or len(matrix[0]) != 2 or len(matrix[1]) != 2:
        return "Matriks harus berukuran 2x2"

    # Menghitung determinan
    determinan = (matrix[0][0] * matrix[1][1]) - (matrix[0][1] * matrix[1][0])
    return determinan

# Contoh penggunaan
A = [[6, 3], [4, 5]]  # Input matriks 2x2
hasil = determinant_2x2(A)
print(f"Determinan = {hasil}")
```
